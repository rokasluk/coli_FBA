```python
#Importing the Flux Scanning based on Enforced Objective Flux function from the cameo package
from cameo.strain_design.deterministic.flux_variability_based import FSEOF

#I believe you could also use cobra for loading the model and it would still work fine
from cameo import load_model

#Pandas for displaying the dataframe of reaction flux changes
import pandas

#To display the full dataframe
pandas.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
#Loading the model
model = load_model('iJO1366_MEV.xml')

#Knocking out the MEP pathway reaction, so only the MEV pathway is used for synthesis
model.reactions.get_by_id("DXPRIi").knock_out()
```

```python
#The MEP model can also be used to compare the efficiency of the pathways
#model = load_model('iJO1366_MEP.xml')
```

```python
#Creating the FSEOF model
fseof = FSEOF(model)

#Running the FSEOF with the beta-carotene sink reaction as the target
a = fseof.run(target=model.reactions.DM_CRTY)
```

```python
#Display of all the reaction flux changes as the flux through the beta-carotene sink reaction increases
df = a.data_frame
df
```
